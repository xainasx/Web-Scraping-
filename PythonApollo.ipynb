# Beautiful Soup

The **BeautifulSoup** library in Python is used for **parsing HTML and XML documents**, making it easy to extract data from web pages. It creates a parse tree from page content, allowing developers to navigate, search, and modify the HTML structure using Pythonic code. BeautifulSoup is especially useful for **web scraping**, where you need to pull specific information—like headlines, links, or tables—from websites. It works well with parsers like `html.parser`, `lxml`, or `html5lib`, and is often used alongside libraries like `requests` to fetch and process web content efficiently.

```
# Import necessary libraries
from bs4 import BeautifulSoup # For parsing HTML
import requests # For making HTTP requests
import pandas as pd # For data manipulation and creating DataFrames
```

```
# Define the URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

# Fetch the content from the URL
page = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page.text, 'html')
```

```
# Print the parsed HTML content
print(soup)
```

```
# Find the first table in the HTML
soup.find('table')
```

# Find Specific Tags or Elements


```
# Find the table with the specified class
soup.find('table', class_ = 'wikitable sortable')
```

```
# Find all tables and select the second one (index 1)
table = soup.find_all('table')[1]
```

```
# Print the selected table
print(table)
```

```
# Find all 'th' (table header) elements within the table
world_titles = table.find_all('th')
```

```
# Display the table header elements
world_titles
```

# Pandas DataFrame


```
# Extract the text from each 'th' element and store in a list to use as DataFrame columns
world_table_titles = [title.text.strip() for title in world_titles]

# Create a Pandas DataFrame with columns based on the extracted table titles
df = pd.DataFrame(columns = world_table_titles)

# Display the empty DataFrame
df
```

```
# Find all 'tr' (table row) elements within the table
column_data = table.find_all('tr')
```

```
# Iterate through the rows (skipping the header row)
for row in column_data[1:]:
    # Find all 'td' (table data) elements in each row
    row_data = row.find_all('td')
    # Extract the text from each 'td' element, strip whitespace, and store in a list
    individual_row_data = [data.text.strip() for data in row_data]

    # Get the current length of the DataFrame
    length = len(df)
    # Add the extracted row data as a new row in the DataFrame
    df.loc[length] = individual_row_data
```

```
# Display the populated DataFrame
display(df)
```

# Saving DataFrame to CSV File


```
# Save the DataFrame to a CSV file named 'Largest companies.csv' in the /tmp directory
# index=False prevents writing the DataFrame index as a column in the CSV
df.to_csv('/tmp/Largest companies.csv', index = False)
```

# Data Analysis and Visualisation

```
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the revenue column is numeric by converting it to a numeric type
df['Revenue (USD billions)'] = pd.to_numeric(df['Revenue (USD billions)'])

# Sort the DataFrame by revenue in descending order for better visualization
df_sorted = df.sort_values('Revenue (USD billions)', ascending=False)

# Create a bar chart to visualize the revenue of the largest companies
plt.figure(figsize=(12, 8)) # Set the figure size for better readability
sns.barplot(x='Revenue (USD billions)', y='Name', data=df_sorted) # Create the bar plot
plt.title('Largest Companies in the US by Revenue') # Set the title of the plot
plt.xlabel('Revenue (USD billions)') # Set the label for the x-axis
plt.ylabel('Company Name') # Set the label for the y-axis
plt.show() # Display the plot
```

```
# Display descriptive statistics for the numerical columns in the DataFrame
# This provides a summary of the central tendency, dispersion, and shape of the distribution
display(df.describe())
```

```
# Create a countplot to visualize the number of companies in each industry
plt.figure(figsize=(10, 8)) # Set the figure size
# Use value_counts().index to order the bars by the frequency of each industry
sns.countplot(y='Industry', data=df_sorted, order = df_sorted['Industry'].value_counts().index)
plt.title('Number of Companies by Industry') # Set the title
plt.xlabel('Number of Companies') # Set the x-axis label
plt.ylabel('Industry') # Set the y-axis label
plt.show() # Display the plot
```

```
# Create a scatter plot to visualize the relationship between the number of employees and revenue
plt.figure(figsize=(10, 6)) # Set the figure size
sns.scatterplot(x='Employees', y='Revenue (USD billions)', data=df) # Create the scatter plot
plt.title('Relationship between Revenue and Employees') # Set the title
plt.xlabel('Number of Employees') # Set the x-axis label
plt.ylabel('Revenue (USD billions)') # Set the y-axis label
plt.show() # Display the plot
```
